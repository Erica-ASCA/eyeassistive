```python
import dlib
import os
import cv2
import numpy as np
```

```python
img_folder = 'frames'
```

```python
for i in os.listdir(img_folder):
    print(i, ":", len(os.listdir(img_folder +"/"+i)))
```

```python
predictor = dlib.shape_predictor('shape_68.dat')
detector = dlib.get_frontal_face_detector()
```

```python
for folder in os.listdir(img_folder):
    for img in os.listdir(img_folder+"/"+folder):
        img_loaded = cv2.imread(img_folder + "/" + folder + "/" + img)
        gray = cv2.cvtColor(img_loaded, cv2.COLOR_BGR2GRAY)
        rects = detector(gray, 1)
        print(rects)
        for (i, rect) in enumerate(rects):
            shape = predictor(gray, rect)
        x_eye_points = [shape.part(x).x for x in range(37,42)]
        y_eye_points = [shape.part(x).y for x in range(37,42)]
        padding = 20
        maxx = max(x_eye_points)
        minx = min(x_eye_points)
        maxy = max(y_eye_points)
        miny = min(y_eye_points)
        crop_image = img_loaded[miny-padding:maxy+padding,minx-padding:maxx+padding]
        if(rects):
            resize = cv2.resize(crop_image, (128,128))
            cv2.imwrite(img_folder + "/" + folder + "/" + "resized " + img, resize)
```
